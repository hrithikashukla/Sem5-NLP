```
import pandas as pd
import re
import numpy as np
import nltk
from gensim.models import Word2Vec
from nltk import word_tokenize
from nltk.corpus import stopwords
import string
nltk.download("stopwords")
```

```
data=pd.read_json("http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Office_Products_5.json.gz",lines=True)
```

```
stop_words = set(stopwords.words('english'))
```

```
def clean_text(text):
  word_filtered = []
  stri = text.translate(str.maketrans('', '', string.punctuation))
  words = text.split()
  for w in words:
    if w not in stop_words:
        word_filtered.append(w)
  final = list(dict.fromkeys(word_filtered))
  return final
```

```
data['unique_words'] = data['reviewText'].apply(clean_text)
data.head()
```

```
words=[[word] for word in data['unique_words'][0]]
```

```
model=Word2Vec(words, size=100, window=5, min_count=1, workers=4)
```

```
for i in words:
  for j in words:
    if j>i:
      print(i[0],' :',j[0],model.similarity(i[0],j[0]))
    else:
      continue
```
